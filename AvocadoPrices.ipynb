```python
region_list=list(avo['region'].unique())
average_price=[]

for i in region_list:
    x=avo[avo.region==i]
    region_average=sum(x.AveragePrice)/len(x)
    average_price.append(region_average)

avo2=pd.DataFrame({'region_list':region_list,'average_price':average_price})
new_index=avo2.average_price.sort_values(ascending=False).index.values
sorted_data=avo2.reindex(new_index)

plt.figure(figsize=(25,12))
sns.barplot(x=sorted_data.region_list,y=sorted_data.average_price)

plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.title('Average Price of Avocado According to Region')
```
